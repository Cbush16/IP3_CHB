### #4

```python
car_sales = pd.read_csv('car-sales.csv')
car_sales.head()
```

```python
car_sales.sample()
```

```python
car_sales.info()
```

```python
car_sales.describe()
```

### #5

```python
car_sales['Price']=car_sales['Price'].replace({'\$':'',',':''},regex=True)
car_sales['Price']=car_sales['Price'].astype('int64')
car_sales['Price']
```

### #6

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
```

```python
car_sales
```

### #7

```python
from datetime import datetime
```

```python
car_sales['Sale Date']=pd.date_range(start=datetime.today(), periods=len(car_sales))
```

```python
car_sales['Sale Date'] = car_sales['Sale Date'].dt.date
```

```python
car_sales
```

### #8

```python
plt.plot(car_sales['Price'],car_sales['Total Sales'])
plt.title('Price vs Total Sales Line Plot')
plt.ylabel('Total Sales')
plt.xlabel('Price')
```

### #9

```python
plt.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
plt.title('Odometer vs Price Scatter Plot')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
```

### #10

```python
plt.bar(car_sales['Make'],car_sales['Odometer (KM)'])
plt.title('Make vs Odometer Bar Plot')
plt.ylabel('Odometer (KM)')
plt.xlabel('Make')
```

### #11

```python
plt.hist(car_sales['Make'])
plt.title('Make Histogrom')
plt.xlabel('Make')
```

### #12

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0, 0].plot(car_sales['Price'], car_sales['Total Sales'])
axes[0, 0].set_title('Price vs Total Sales Line Plot')
axes[0, 0].set_xlabel('Price')
axes[0, 0].set_ylabel('Total Sales')

axes[0, 1].scatter(car_sales['Odometer (KM)'], car_sales['Price'])
axes[0, 1].set_title('Odometer (KM) vs Price Scatter Plot')
axes[0, 1].set_xlabel('Odometer (KM)')
axes[0, 1].set_ylabel('Price')

axes[1, 0].bar(car_sales['Make'], car_sales['Odometer (KM)'])
axes[1, 0].set_title('Make vs Odometer Bar Plot')
axes[1, 0].set_ylabel('Odometer (KM)')
axes[1, 0].set_xlabel('Make')

axes[1, 1].hist(car_sales['Make'])
axes[1, 1].set_title('Make Histogram')
axes[1, 1].set_xlabel('Make')
plt.show()
```

### #13

```python
sns.boxplot(x='Colour',y='Total Sales',data=car_sales)
```

### #14

```python
plt.figure(figsize=(6,4))
sns.lmplot(x='Total Sales', y='Odometer (KM)', hue='Make', ci=None ,data=car_sales)
plt.ylim(0, 250000)
plt.xlim(0,78000)
plt.show()
```

### #16

```python
corr_matrix=car_sales[['Total Sales','Price','Odometer (KM)']].corr()
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

Price and Odometer have a relatively strong negative correlation, Price and Total Sales have a weak positive correlation, Total Sales and Odometer have weak negative correlation.
